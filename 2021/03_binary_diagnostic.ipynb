```python
# import data
#filename = '03_test.txt'
filename = '03_data.txt'
raw = open(f'./data/{filename}').read()
```

```python
import numpy as np
import pandas as pd
import itertools

# define data process function
def process(data):
    data = data.split('\n')
    data = [list(x) for x in data]
    return data

# process data
data = process(raw)
```

```python
def calculate_power(data):
    df = pd.DataFrame(data)
    gamma = []
    epsilon = []
    for column in df:
        gamma.append(df[column].value_counts().idxmax())
        epsilon.append(df[column].value_counts().idxmin())
    gamma_dec = int("".join([i for i in gamma]), 2)
    epsilon_dec = int("".join([i for i in epsilon]), 2)
    return gamma_dec*epsilon_dec

calculate_power(data)
```

### Part 2

```python
ndata = data.copy()
ncols = len(data[0])

for c_id in range(ncols):
    a = [row[c_id] for row in ndata]
    values, counts = np.unique(a, return_counts=True)
    most_common = values[np.argmax(counts)]
    if counts[0] != counts[1]:
        ndata = [row for row in ndata if row[c_id] == most_common]
    else:  # tie case
        ndata = [row for row in ndata if row[c_id] == "1"] 
o2 = list(itertools.chain(*ndata))
o2 = int("".join([i for i in o2]), 2)
o2
```

```python
ndata = data.copy()
ncols = len(data[0])

for c_id in range(ncols):
    a = [row[c_id] for row in ndata]
    values, counts = np.unique(a, return_counts=True)
    least_common = values[np.argmin(counts)]
    if len(counts) > 1:
        if counts[0] != counts[1]:
            ndata = [row for row in ndata if row[c_id] == least_common]
        else:  # tie case
            ndata = [row for row in ndata if row[c_id] == "0"]
    else:
        pass
co2 = list(itertools.chain(*ndata))
co2 = int("".join([i for i in co2]), 2)
```

```python
ans = o2*co2
ans
```
